## Random Graph

```python
import networkx as nx
import random
```

```python
n_nodes = 20
p_connect = 0.3

g = nx.Graph()
g.add_nodes_from(range(n_nodes))

for i in range(n_nodes):
  for j in range(i+1,n_nodes):
    if random.random() < p_connect:

      g.add_edge(i,j)
nx.draw(g,with_labels = True)
```

## Small World Graph


```python
n_nodes = 20
k_neigh = 4
p_wire = 0.3

g = nx.watts_strogatz_graph(n_nodes,k_neigh,0)
nx.draw(g,with_labels = True)
```

```python
g = nx.random_regular_graph(k_neigh,n_nodes)
for u,v in g.edges():
  if type(v) == 'int' and (u < v) and (random.random() <= p_wire):
    w = random.choice(list(g.edges()))

    if(u,w) not in g.edges() and (v,w) not in g.edges():

      g.remove_edge(u,v)
      g.add_edge(u,w)
      g.add_edge(v,w)

nx.draw(g,with_labels = True)
```

```python
print("CC:",nx.average_clustering(g))
```

## Preferential Attachment Model

```python
n_nodes =40
m =4
g = nx.barabasi_albert_graph(n_nodes, m)
nx.draw(g,with_labels = True)
```

```python
highest_degree_node = max(dict(g.degree()).items(), key = lambda k: k[1])[0]
```

```python
new_node = 40
g.add_node(new_node)
g.add_edge(highest_degree_node, new_node)
node_c = ['red' if node == 40 else 'blue' for node in g.nodes()]
nx.draw(g,node_color = node_c)
```
